```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('./world-development-indicators/Indicators.csv')
```

```python
data.tail()
```

```python
indicator='Arms import'
```

```python
country='India'
```

```python
mask1=data['IndicatorName'].str.contains(indicator)
```

```python
mask2=data['CountryName'].str.contains(country)
```

```python
result=data[mask1 & mask2]
```

```python
result
```

```python
years=result['Year'].values
```

```python
values=result['Value'].values
```

```python
plt.plot(years,values)
plt.xlabel('Years')
plt.ylabel('Value')
plt.title('Arms Import Cost for India')
```

```python
plt.show()
```
