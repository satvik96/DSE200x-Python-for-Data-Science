```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('./weather/daily_weather.csv')
```

```python
del data['number']
```

```python
data=data.dropna()
```

```python
clean_data=data.copy()
clean_data['High Humidity label']=(clean_data['relative_humidity_3pm']>24.99)*1
```

```python
print(clean_data['High Humidity label'])
```

```python
y=clean_data[['High Humidity label']].copy()
y.head()
```

```python
clean_data['relative_humidity_3pm'].head()
```

```python
morning_features=['air_pressure_9am','air_temp_9am','avg_wind_direction_9am','avg_wind_speed_9am','max_wind_direction_9am','max_wind_speed_9am','rain_accumulation_9am','rain_duration_9am','relative_humidity_9am']
```

```python
x=clean_data[morning_features].copy()
```

```python
x
```

```python
y
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=324)
```

```python
type(x_train)
```

```python
type(x_test)
```

```python
type(y_train)
```

```python
type(y_test)
```

```python
x_train.describe()
```

```python
humidity_classifier=DecisionTreeClassifier(max_leaf_nodes=10 , random_state=0)
humidity_classifier.fit(x_train,y_train)
```

```python
type(humidity_classifier)
```

```python
predictions=humidity_classifier.predict(x_test)
```

```python
predictions[:10]
```

```python
y['High Humidity label'][:10]
```

```python
accuracy_score(y_true=y_test,y_pred=predictions)
```
